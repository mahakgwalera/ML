```python
# Types of Graphs
1. plot-    line graph, between two parameters
2. bar-     bar graph of single parameter
3. hist-    histogram  of a bunch of data i.e. frequency of single parameter
4. scatter-  dot graph between two parameters(params)
5. boxplot/box- statistical overview in form of box for a asinglw parameter(param)-mean,outliers
, quantiles(25%, 75%), IQR (optional-categorical data)
6. stem-  more like bar graph, Lines instead of bars
7. violinplot-  violin shaped statistical overview graph spread of data (more like histogram) single parameter 
(optional catagorical data)
```

```python
# Types of Graphs
1. plot-    line graph, between two parameters
2. bar-     bar graph of single parameter
3. hist-    histogram  of a bunch of data i.e. frequency of single parameter
4. scatter-  dot graph between two parameters(params)
5. boxplot/box- statistical overview in form of box for a asinglw parameter(param)-mean,outliers
, quantiles(25%, 75%), IQR (optional-categorical data)
6. stem-  more like bar graph, Lines instead of bars
7. violinplot-  violin shaped statistical overview graph spread of data (more like histogram) single parameter 
(optional catagorical data)
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
x1=np.arange(1,11,3)
```

```python
y=x**3
```

```python
plt.plot(x,y)
```

```python
plt.bar(x,y)
plt.title('bar graph')
plt.grid()
plt.xlabel('x-->')
plt.ylabel('y-->')
plt.xticks(x)
plt.yticks(range(0,200,10))
plt.show()
```

```python
plt.hist(x,y)
plt.title('histogram')
plt.grid()
plt.xlabel('x-->')
plt.ylabel('y-->')
plt.xticks(x)
plt.yticks(range(0,200,10))
plt.show()
```

```python
plt.scatter(x,y)
plt.title('scatter')
plt.grid()
plt.xlabel('x-->')
plt.ylabel('y-->')
plt.xticks(x)
plt.yticks(range(0,200,10))
plt.show()
```

```python

plt.stem(x,y)
plt.title('stem')
plt.grid()
plt.xlabel('x-->')
plt.ylabel('y-->')
plt.xticks(x)
plt.yticks(range(0,200,10))
plt.show()
```

```python
y2=x**4
```

```python
plt.plot(x,y2)
plt.show()
```

```python
plt.plot(x,y,x,y2)
plt.legend(['y=**3', 'y2=x**4'])
plt.show()
```

```python
fig=plt.figure()
plt.plot(x,y,lw=3, marker='o',markerfacecolor='yellow')
plt.plot(x,y2, marker='v', markersize=15)
plt.legend(['y=x**3','y2=x**4'])
plt.grid()
plt.show()
```
